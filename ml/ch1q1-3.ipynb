# 今日任务：低维数据可视化 I


安装软件 python 和工具包 sklearn


```python
# 导入 sklearn 第三方库
from sklearn.datasets import make_circles, make_moons
import matplotlib.pyplot as plt
```

生成数据 make_circles 和 make_moons，并显示 X = 400 × 2, Y = {0, 1}^400 ，画图： 


```python

# 生成 circles 数据
circles_x, circles_y = make_circles(n_samples=400, noise=0.1, factor=0.2)

# 生成 moons 数据
moons_x, moons_y = make_moons(n_samples=400, noise=0.05)
```

```python

colors = ['red', 'blue']

# 绘制 circles 数据
plt.figure(figsize=(4, 4))
for i in range(2):
    plt.scatter(circles_x[circles_y == i, 0], circles_x[circles_y == i, 1], c=colors[i], label=f'Class {i}')
plt.title("Data by Make_circles()")
plt.legend(loc='lower right')
plt.show()

# 绘制 moons 数据
plt.figure(figsize=(4, 4))
for i in range(2):
    plt.scatter(moons_x[moons_y == i, 0], moons_x[moons_y == i, 1], c=colors[i], label=f"Class {i}")
plt.title("Data by Make_moons()")
plt.legend(loc='lower right')
plt.show()
```

高维数据如何可视化？

答：可以用 t-SNE、UMAP 或者用 PCA 技术进行降维，然后再使用 matplotlib 或者其他可视化库进行 2D、3D 可视化。
